```python
class MLP(torch.nn.Module):
  def __init__(self, inp_size, hidden_size):
    super().__init__()
    self.inp_size = inp_size
    self.hidden_size = hidden_size
    self.mlp = torch.nn.Sequential(
        torch.nn.Linear(inp_size, hidden_size),
        torch.nn.ReLU(),
        torch.nn.Linear(hidden_size, inp_size)
    )

  def forward(self, x):
    return self.mlp(x)
```

```python
class ModelWrapper(torch.nn.Module):
  def __init__(self, model, tokenizer, hidden_size, num_tokens, temperature):
    super().__init__()
    self.model = model
    self.tokenizer = tokenizer
    # self.clip_model = clip_model
    self.hidden_size = hidden_size
    self.num_tokens = num_tokens
    self.temperature = temperature
    self.mlp = MLP(self.model.config.hidden_size, hidden_size)
    self.gumbel = GumbelSampler(self.tokenizer, self.temperature)
    # self.mlps = [MLP(self.model.config.hidden_size, hidden_size) for _ in range(self.num_tokens)]
    # self.gumbels = [GumbelSampler(self.tokenizer, self.temperature) for _ in range(self.num_tokens)]

  def forward(self, inputs):
#     print('2'*10)
    tokens = self.tokenizer(inputs, padding='max_length', max_length=2, add_special_tokens=False, return_tensors='pt')
#     print('tokens: ',tokens)
    hidden_states = self.model.bert(**tokens)[0]
#     print('hidden_states: ', hidden_states.shape)
    logits = self.mlp(hidden_states)
#     print('logits: ',logits.shape)
    logits = self.model.cls(logits)
#     print('logits: ',logits.shape)
    clip_embed, txt_tokens = self.gumbel(logits, inputs)
#     print('clip_embed: ', clip_embed, clip_embed.shape)
    return clip_embed, txt_tokens
```

```python
class WholeModel(torch.nn.Module):
  def __init__(self, wrapper, clip_model):
    super().__init__()
    self.wrapper = wrapper
    self.clip_model = clip_model
    self.clip_tokenizer = AutoTokenizer.from_pretrained("openai/clip-vit-base-patch32")

  def forward(self, inputs):
    self.embed, feed = self.wrapper(inputs)
#     print('faile here: ', self.embed[0,:,:])

    x = self.embed + self.clip_model.positional_embedding.type(clip_model.dtype)
    x = x.permute(1, 0, 2)  # NLD -> LND
    x = clip_model.transformer(x)
    x = x.permute(1, 0, 2)  # LND -> NLD
    x = clip_model.ln_final(x).type(clip_model.dtype)
    x = x[torch.arange(x.shape[0]), feed.argmax(dim=-1)] @ clip_model.text_projection
    self.txt_encodings = x.float() / x.norm(dim=-1, keepdim=True, dtype=torch.float)
    return self.txt_encodings
#     self.img_encodings = torch.randn((3, 512), dtype=torch.float).to(device)
#     sim = self.img_encodings @ self.txt_encodings.T
#     sim_ = torch.softmax(sim, dim=-1)
#     print(sim_)
#     class_p = torch.diagonal(sim_).mean()
#     print('class probabilities: ', torch.diagonal(sim_))
#     eye_inv = (1 - torch.eye(sim_.shape[0], sim_.shape[1])).to(device)
#     sim_off = sim_ * eye_inv
#     print('other class maxima: ', torch.max(sim_off, dim=-1))
#     non_class_p = torch.max(sim_off, dim=-1).values.mean()
#     print('class p: ', class_p)
#     print('non class p: ', non_class_p)
#     loss = class_p - non_class_p
#     return loss
```

```python
def loss_func(img_encodings, txt_encodings):
    sim = img_encodings @ txt_encodings.T
    sim_ = torch.softmax(sim, dim=-1)
    print('sim score: ', sim_)
    class_p = torch.diagonal(sim_)
    eye_inv = (1 - torch.eye(sim_.shape[0], sim_.shape[1])).to(device)
    sim_off = sim_ * eye_inv
    non_class_p = torch.max(sim_off, dim=-1).values
    loss = torch.max(1+non_class_p-class_p, torch.zeros_like(class_p))
    loss = loss.mean()
    return loss
```

```python
def train(train_model, dataloader, num_epochs):
    train_model.train()
    for i in range(num_epochs):
        print('Epoch: ', i+1)
        running_loss = 0.
        running_size = 0
        for j, data in enumerate(dataloader):
            # Every data instance is an input + label pair
#             img_features = clip_model.encode_image(data.to(device))
            #################################
            optimizer.zero_grad()
            txt_embeddings = train_model(inputs)
            img_encodings = clip_model.encode_image(data.to(device)).float() #clip image embedding
            img_encodings /= img_encodings.norm(dim=-1, keepdim=True)
            loss = loss_func(img_encodings, txt_embeddings)
            ################################
            loss.backward()
            optimizer.step()

            # Gather data and report
            running_loss += loss.item()*data.size(0)
            running_size += data.size(0)
            print('running_loss: ', running_loss / running_size)

        print('running loss: ', running_loss / running_size)
```

```python
class CocoDataset():
    def __init__(self, img_ids, bboxes, img_path):
        self.img_ids  = img_ids
        self.bboxes = bboxes
        self.img_path = img_path
        self.prefix = '000000000000'
        self.len = len(self.img_ids)

    def __len__(self):
        return self.len

    def __getitem__(self, idx):
        img_id = self.img_ids[idx]
        img_name = self.prefix[0:12-len(str(img_id))] + str(img_id) + '.jpg'
        img = np.asarray(Image.open(self.img_path + 'COCO_val2014_' + img_name))
        x, y, w, h = int(self.bboxes[idx][0]), int(self.bboxes[idx][1]), int(self.bboxes[idx][2]), int(self.bboxes[idx][3])
#         print(x,y,w,h)
        w = w+1 if w==0 else w
        h = h+1 if h==0 else h
        img = processor(Image.fromarray(img[y:y+h,x:x+w]))
        return img
```
